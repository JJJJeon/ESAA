## matplotlib 라이브러리 자유자재로 사용하기

```python
tips=sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

히스토그램 - 데이터프레임의 열 데이터 분포와 빈도를 살펴보는 용도로 자주 사용, 일변량 그래프

```python
fig=plt.figure()
axes1=fig.add_subplot(1,1,1) #기본 틀과 격자
```

```python
axes1.hist(tips['total_bill'],bins=10) #x축의 간격=10
axes1.set_title("Histogram of Total Bill")
axes1.set_xlabel('Frequency')
axes1.set_ylabel("Total Bill")

fig
```

산점도 그래프 - 이변량 그래프

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

박스 그래프 - 이산형 변수와 연속형 변수를 함께 사용하는 그래프

```python
boxplot=plt.figure()
axes1=boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex']=='Female']['tip'],
               tips[tips['sex']=='Male']['tip']],
              labels=['Female','Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

디변량 그래프

```python
#성별을 색상으로 구분하기 위해 문자를 정수로 치환하는 함수
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
```

```python
tips['sex_color']=tips['sex'].apply(recode_sex) #함숫값을 새로운 열에 추가
```

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c=tips['sex_color'],
    alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## seaborn 라이브러리 자유자재로 사용하기

```python
import seaborn as sns

tips=sns.load_dataset('tips')
ax=plt.subplots() #기본 틀
ax=sns.distplot(tips['total_bill']) #히스토그램 - 히스토그램과 밀집도 그래프 같이 그림
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],kde=False) #밀집도 그래프 제외
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],hist=False) #밀집도 그래프만
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],rug=True) #양탄자 그래프
ax.set_title('Total Bill Histogram with Density and Rug plot')
ax.set_xlabel('Total Bill')
```

count 그래프

```python
ax=plt.subplots()
ax=sns.countplot('day',data=tips)
ax.set_title('Cound of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

다양한 종류의 이변량 그래프

```python
ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips) #회귀선
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips, fit_reg=False) #회귀선 제거
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
joint=sns.jointplot(x='total_bill',y='tip',data=tips) #산점도와 히스토그램 한번에
joint.set_axis_labels(xlabel='Total Bill',ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip',fontsize=10,y=1.03)
```

```python
hexbin=sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') #육각 그래프
hexbin.set_axis_labels(xlabel='Total Bill',ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip',fontsize=10,y=1.03)
```

```python
ax=plt.subplots()
ax=sns.kdeplot(data=tips['total_bill'], #이차원 밀집도
               data2=tips['tip'],
               shade=True) #음영효과
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax=plt.subplots()
ax=sns.barplot(x='time',y='total_bill',data=tips) #바 그래프: 지정한 변수의 평균 계산하여 그림
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
ax=plt.subplots()
ax=sns.boxplot(x='time',y='total_bill',data=tips) #박스 그래프
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total bill')
```

```python
ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',data=tips) #바이올린 그래프: 박스 그래프에 커널 밀도 추정
ax.set_title('Violin plot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total bill')
```

```python
fig=sns.pairplot(tips) #관계 그래프
```

```python
pair_grid=sns.PairGrid(tips)
pair_grid=pair_grid.map_upper(sns.regplot) #대각선 기준 위쪽에 그릴 그래프 지정
pair_grid=pair_grid.map_lower(sns.kdeplot) #대각선 기준 아래쪽에 그릴 그래프 지정
pair_grid=pair_grid.map_diag(sns.distplot,rug=True)
plt.show()
```

```python
ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True) #바이올린에 색상 추가
```

```python
scatter=sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',fit_reg=False) #산점도에 색상 추가
```

```python
fig=sns.pairplot(tips,hue='sex') #관계 그래프에 색상 추가
```

```python
scatter=sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False, hue='sex',
                   scatter_kws={'s':tips['size']*10}) #산점도 점 크기 조절
```

```python
scatter=sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False,hue='sex',
                   markers=['o','x'],scatter_kws={'s':tips['size']*10}) #산점도 점 다른 기호로 표현
```

```python
anscombe_plot=sns.lmplot(x='x',y='y',data=anscombe,fit_reg=False,
                         col='dataset',col_wrap=2) #그룹별로 그래프, col_wrap:그래프 그릴 열 최댓값 지정, col: 데이터 그룹 구분할 열 지정
```

## seaborn 라이브러리로 그래프 스타일 설정하기

```python
fig,ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

```python
sns.set_style('whitegrid')
fig,ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True) #whitegrid
```

```python
fig=plt.figure()

seaborn_styles=['darkgrid','whitegrid','dark','white','ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position=idx+1
  with sns.axes_style(style):
    ax=fig.add_subplot(2,3,plot_position)
    violin=sns.violinplot(x='time',y='total_bill',data=tips,ax=ax)
    violin.set_title(style)

fig.tight_layout() #모든 스타일 하나씩 적용
```
